### Montando um Classificador Naive-Bayes

Considerando apenas as mensagens da planilha Treinamento, ensine  seu classificador.

```python
irrelevante = train.loc[train["AvaliaÃ§Ã£o"]==0,"Treinamento"]
tweet_lista = train["Treinamento"].tolist()
tweet_lista
```

```python
import re
def lista_para_string(s):
    str1 = " "
    return (str1.join(s))

x = lista_para_string(tweet_lista)
x
```

```python
def cleanup(text):
    punctuation = '[!-.:?;@/"\,]'
    pattern = re.compile(punctuation)
    text_subbed = re.sub(pattern, '', text)
    return text_subbed

texto_limpo_irrelevante =  cleanup(x.lower())

todas_palavras_irrelevante = texto_limpo_irrelevante.split()

serie_irrelevante = pd.Series(todas_palavras_irrelevante)
serie_irrelevante
```

```python
relevante= train.loc[train["AvaliaÃ§Ã£o"]==1,"Treinamento"]
tweet_lista = relevante.tolist()

import re 
def lista_para_string(s):  
    str1 = " " 
    return (str1.join(s)) 

y = lista_para_string(tweet_lista)

def cleanup(text):
    punctuation = '[!-.:?;@/"\,]'
    pattern = re.compile(punctuation)
    text_subbed = re.sub(pattern, '', text)
    return text_subbed

texto_limpo_relevante =  cleanup(y.lower())

todas_palavras_relevante = texto_limpo_relevante.split()

serie_relevante = pd.Series(todas_palavras_relevante)
serie_relevante
```

```python
muito_relevante = train.loc[train["AvaliaÃ§Ã£o"]==2,"Treinamento"]
tweet_lista = muito_relevante.tolist()

import re
def lista_para_string(s):
    str1 = " "
    return (str1.join(s))

z = lista_para_string(tweet_lista)

def cleanup(text):
    punctuation = '[!-.:?;@/"\,]'
    pattern = re.compile(punctuation)
    text_subbed = re.sub(pattern, '', text)
    return text_subbed

texto_limpo_muito_relevante =  cleanup(z.lower())

todas_palavras_muito_relevante = texto_limpo_muito_relevante.split()

serie_muito_relevante = pd.Series(todas_palavras_muito_relevante)
serie_muito_relevante
```

```python
tabela_irrelevante_absoluta = serie_irrelevante.value_counts()
print(tabela_irrelevante_absoluta)
```

```python
tabela_relevante_absoluta = serie_relevante.value_counts()
print(tabela_relevante_absoluta)
```

```python
tabela_muito_relevante_absoluta = serie_muito_relevante.value_counts()
print(tabela_muito_relevante_absoluta)
```

```python
tudo = texto_limpo_irrelevante + texto_limpo_relevante + texto_limpo_muito_relevante
serie_tudo = pd.Series(tudo.split())
tabela_serie_tudo_absoluta = serie_tudo.value_counts()
print(tabela_serie_tudo_absoluta)
```

```python
ProbI = len(serie_irrelevante)/len(serie_tudo)
ProbR = len(serie_relevante)/len(serie_tudo)
ProbMR = len(serie_muito_relevante)/len(serie_tudo)
print(ProbI, ProbR, ProbMR)
```

```python
for f in test.Teste:
    frase = f
    frase = cleanup(frase.lower())
    frase_split = frase.split()

    ProbIrrelevante = 1
    ProbRelevante = 1
    ProbMuitoRelevante = 1
    for i in frase_split:
        if i in tabela_irrelevante_absoluta:
            ProbIrrelevante *= (tabela_irrelevante_absoluta[i]+1)/(len(serie_irrelevante) + len(tabela_serie_tudo_absoluta))
        if i in tabela_relevante_absoluta:
            ProbRelevante *= (tabela_relevante_absoluta[i]+1)/(len(serie_relevante) + len(tabela_serie_tudo_absoluta) )
        if i in tabela_muito_relevante_absoluta:
            ProbMuitoRelevante *= (tabela_muito_relevante_absoluta[i]+1)/(len(tabela_serie_tudo_absoluta) + len(serie_muito_relevante))
        if i not in tabela_irrelevante_absoluta:
            ProbIrrelevante *= 1/(len(serie_irrelevante) + len(tabela_serie_tudo_absoluta))
        if i not in tabela_relevante_absoluta:
            ProbRelevante *= 1/(len(serie_relevante) + len(tabela_serie_tudo_absoluta))
        if i not in tabela_muito_relevante_absoluta:
            ProbMuitoRelevante *= 1/(len(serie_muito_relevante) + len(tabela_serie_tudo_absoluta))
                

    if (ProbIrrelevante < ProbMuitoRelevante) and (ProbRelevante < ProbMuitoRelevante):
        test.loc[(test.Teste==f),'Relevancia'] = 'Muito Relevante'
    if (ProbRelevante > ProbIrrelevante) and (ProbRelevante > ProbMuitoRelevante):
        test.loc[(test.Teste==f),'Relevancia'] = 'Relevante'
    if (ProbRelevante < ProbIrrelevante) and (ProbIrrelevante > ProbMuitoRelevante):
        test.loc[(test.Teste==f),'Relevancia'] = 'Irrelevante'
        
test.head(20)
```
